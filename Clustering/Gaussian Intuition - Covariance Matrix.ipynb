```python
from scipy.stats import multivariate_normal
import numpy as np
import plotly.graph_objects as go
```

```python
x = np.linspace(-2,6,100)
y = np.linspace(-2,6,100)
```

```python
grid = np.meshgrid(x,y)
```

```python
cov = np.diag(np.ones(2))
mean = np.ones(2)*2
z = multivariate_normal.pdf(np.dstack(grid).reshape(-1,2), mean=mean,cov=cov).reshape(100,100)
```

```python
fig = go.Figure(data=[go.Surface(z=z,x=x,y=y)])
fig.update_traces(contours_z=dict(show=True, usecolormap=True,
                                  highlightcolor="limegreen", project_z=True))
fig.update_layout(title='Gaussian', autosize=False,
                  width=900, height=800,
                  margin=dict(l=65, r=50, b=65, t=90))

fig.show()
```

```python
cov = np.array([[1,0],[0,2]]) #increasing the variance in the y direction
mean = np.ones(2)*2
z = multivariate_normal.pdf(np.dstack(grid).reshape(-1,2), mean=mean,cov=cov).reshape(100,100)
```

```python
fig = go.Figure(data=[go.Surface(z=z,x=x,y=y)])
fig.update_traces(contours_z=dict(show=True, usecolormap=True,
                                  highlightcolor="limegreen", project_z=True))
fig.update_layout(title='Gaussian', autosize=False,
                  width=900, height=800,
                  margin=dict(l=65, r=50, b=65, t=90))

fig.show()
```

```python
cov = np.array([[1,0],[0,2]]) #increasing the variance in the y direction

theta = np.pi/4
R = np.array([[np.cos(theta), np.sin(theta)],[-np.sin(theta),np.cos(theta)]])
rotated_cov = R.dot(cov).dot(R.T) # rotate the cov matrix by 45 degrees

mean = np.ones(2)*2
z = multivariate_normal.pdf(np.dstack(grid).reshape(-1,2), mean=mean,cov=rotated_cov).reshape(100,100)
```

```python
fig = go.Figure(data=[go.Surface(z=z,x=x,y=y)])
fig.update_traces(contours_z=dict(show=True, usecolormap=True,
                                  highlightcolor="limegreen", project_z=True))
fig.update_layout(title='Gaussian', autosize=False,
                  width=900, height=800,
                  margin=dict(l=65, r=50, b=65, t=90))

fig.show()
```
